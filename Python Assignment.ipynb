```python
def show_menu():
    print("\nMENU")
    print("1. View Record")
    print("2. Health Risk")
    print("3. Exit")
```

```python
show_menu()
```

```python
while True:
    show_menu()
    choice = input("Enter choice: ")

    if choice == "1":
        print("You selected View Record")
    elif choice == "2":
        print("You selected Health Risk")
    elif choice == "3":
        print("Program Ended")
        break
    else:
        print("Invalid choice")
```

```python
def view_records():
    print("First Record:")
    print(records[0])

def health_risk_analysis():
    cal = records[0]["Average_Daily_Calories"]

    if cal > 2500:
        print("Health Risk: High Risk")
    elif cal > 1800:
        print("Health Risk: Moderate Risk")
    else:
        print("Health Risk: Low Risk")
```

```python
view_records()
```

```python
health_risk_analysis()
```
