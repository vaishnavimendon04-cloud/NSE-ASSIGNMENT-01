```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Identify column types
text_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include='number').columns
```

```python
for col in text_cols:
    df[col] = df[col].str.strip()

# 3. Text casing standardization
for col in text_cols:
    df[col] = df[col].str.upper()
```

```python
if 'year' in df.columns:
    df['year'] = pd.to_datetime(df['year'], format='%Y', errors='coerce')
```

```python
for blood_glucose_level in num_cols:
    df[blood_glucose_level] = df[blood_glucose_level].fillna(df[blood_glucose_level].median())
```

```python
for smoking_history	 in text_cols:
    df[smoking_history	] = df[smoking_history	].fillna(df[smoking_history	].mode()[0])
```

```python
df = df.drop_duplicates()
```

```python
df['bmi_glucose_diff'] =  df['blood_glucose_level']-df['bmi']
df['bmi_glucose_diff']
```

```python
df['health_status'] = df.apply(lambda x: 'HIGH RISK'if (x['bmi'] >= 30) and (x['blood_glucose_level'] >= 140) and (x['hba1c_level'] >= 6.5)
    else 'LOW RISK',
    axis=1)
df['health_status']
```

```python
df['overall_high_risk'] = ((df['bmi'] > df['bmi'].mean()) &(df['blood_glucose_level'] > df['blood_glucose_level'].mean()) &(df['hba1c_level'] > df['hba1c_level'].mean()))
df['overall_high_risk']
```

```python
print("\nINSIGHT 1: Average Blood Glucose by Health Status")
print(
    df.groupby('health_status')['blood_glucose_level']
      .mean()
      .sort_values(ascending=False)
)
```

```python
print("\nINSIGHT 2: Average HbA1c by Gender")
print(
    df.groupby('gender')['hba1c_level']
      .mean()
      .sort_values(ascending=False)
)
```
